```python
# 누락값 처리

# 0값으로 대체
# 평균값으로 대체
# 해당 행 삭제
# 주변 값으로 대체
```

```python
ebola.fillna(0) # NaN값에 모두 0 넣기
```

```python
# 평균값 넣기
ebola['Cases_Guinea'].fillna(ebola['Cases_Guinea'].mean())
```

```python
# 주변값 넣기 -> 바로 전 값과 그 다음 값의 평균을 계산해서 넣어줌

ebola = ebola.interpolate() # 앞에 있는 데이터, 뒤의 데이터의 평균값으로 누락값 처리해주는 함수
ebola
```

```python
# 하지만 앞,뒤의 데이터도 NaN이라면 여전히 NaN
# 그래서, 그냥 뒤에 있는 값으로 똑같이 채움

ebola = ebola.fillna(method='bfill')  # back fill
ebola
```

```python
ebola = ebola.fillna(method='ffill')  # front fill
ebola
```
