```python
# 행을 열로 올리기

weather = pd.read_csv('weather.csv')
weather

# tmax는 월별 최고온도가 있던 날시를 표시
# tmin는 월별 최저온도가 있던 날짜를 표시
```

```python
weather2 = pd.melt(weather, id_vars=['id','year','month','element'], var_name='day', value_name='temp')
weather2

# element열의 tmax, tmin를 열로 올리면 좋을 것 같음
```

```python
weather3 = weather2.pivot_table(
    index = ['id','year','month','day'], # 열로 올리지않고 고정시킬 열
    columns = 'element', # 열로 올리고 싶은 값을 갖고있는 현재 열
    values = 'temp',  # 새로 올린 열에 들어갈 값
    dropna = False  # True로 설정하면, None값인 행들 삭제함
)
weather3

# tmax, tmin의 출처가 element에서 왔다고 표시해줌
# 행의 값을 열로 올렸더니 day 열의 값들의 순서가 섞임
```

```python
weather3.reset_index()  # 깔끔하게 정리해줌

# 왼쪽 끝의 element는 열 이름 아님 (데이터프레임 이름)
```

```python
# 열 이름을 아는 경우에는 names 매개변수를 사용해서 지정할 수 있다.

df = pd.read_csv('pima-indians-diabetes.csv',
                 names = ['pregnant','plasma','pressure','thickness','insulin','BMI','pedigree','age','class'])
df

# pregnant : 임신횟수, plasma : 혈당 농도, pressure : 혈압, thickness : 지방 두께
# insulin : 인슐린, BMI : 체질량 지수, pedigree : 당뇨병 가족력, age : 나이
# class : 1 이면 당뇨, 0 이면 정상
```

```python
# 임신횟수와 당뇨병/비만과의 영향
# -> 임신횟수별로 당뇨의 평균값

a = df.groupby('pregnant')['class'].mean()
a
```

```python
a.plot()

# 임신횟수가 증가할수록 당뇨병 평균값이 올라감
```

```python
# 상관계수 구하기
# : 두 개의 변수 사이의 상관관계, 모든 데이터를 대표하는 하나의 직선을 그렸을 때의 기울기 값
# 상관계수 값이 1에 가까울수록 완벽한 상관관계가 있고, -1에 가까울수록 반비례하는 상관관계가 있다.
# 상관계수 값이 0일때 아무런 관계도 없음을 의미한다.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (12,12))
sns.heatmap(df.corr(), linecolor='white', annot=True) # .corr()가 상관계수 함수
plt.show()
```

```python
# class-pregnant 의 상관계수 = 0.22 -> 높다고 할 수 없지만 그렇다고 낮다고 할 수도 없음
# class-plasma 의 상관계수 = 0.47
```

```python
# 퀴즈 : 당뇨별로 plasma의 히스토그램
# FacetGrid : 집단별로 그림을 그릴 때 사용

facet = sns.FacetGrid(df, col='class')
facet.map(plt.hist, 'plasma', bins=10)
plt.show()
```

```python
# 당뇨병일때와 정상일때 히스토그램의 큰 차이점은 plasma값이 150 이상인 부분이다.
# plasma 값이 150이상 사람은 당뇨병일 확률이 높음
```

```python
# ----------------------------------------------------

### 데이터 크롤링 : 인터넷 데이터를 가져오는 기술

# 크롤링 vs 스크래핑
# 크롤링 : 페이지의 정보, 데이터를 전부 가져오는 것
# 스크래핑 : 부분적으로 가져오는 것

# 인터넷은 크게봤을 때 HTML언어로 구성되어있음

import requests   # 데이터 요청을 위한 라이브러리
from bs4 import BeautifulSoup   # 파이썬으로 HTML을 다루는 라이브러리
import pandas as pd   # 엑셀화
import time   # 데이터를 천천히 가져가기 위해서 참조

# 977회차 로또
url = requests.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%A1%9C%EB%98%90')
url
```

```python
# Response 200 : 정상
# 400대 : 없는 페이지
# 500대 : 문제가 있거나 거절당함
```

```python
url.text  # 해당 주소 크롤링 -> 문자열 데이터, html아님
```

```python
html = BeautifulSoup(url.text)  # BeautifulSoup 함수를 사용해서 해당 문자열을 HTML로 번역
html
```

```python
# 크롬환경에서 로또번호에 커서를 두고 우클릭 후 검사를 누르면 로또번호에 대한 html 코드를 볼 수 있다.
# 로또번호는 div 태그의 class = 'num_box' 안에 들어있는 것을 확인할 수 있다.

html.find('div', class_ = 'num_box')  # 로또번호에 대한 부분만 찾아옴
```

```python
# 더 자세하게 이어서 찾아나갈 수 있음

numbers = html.find('div', class_ = 'num_box').find_all('span')
numbers
```

```python
# 주의할 점
# find() : HTML로 반환
# find_all() : 리스트로 반환

# 따라서 find_all() 뒤에 다시 .find() 사용이 불가하다.
```

```python
# '977회차' 도 크롤링해오기

html.find('a', class_ = '_lotto-btn-current').find('em').text
```

```python
# '회' 빈칸처리
html.find('a', class_ = '_lotto-btn-current').find('em').text.replace('회', '')
```

```python
current = int(html.find('a', class_ = '_lotto-btn-current').find('em').text.replace('회', ''))
current
```

```python
# 1회차부터 977회차까지의 로또번호 전부 크롤링해오기

for n in range(1, current+1):
  url = requests.get(f'https://search.naver.com/search.naver?where=nexearch&sm=top_sly.hst&fbm=1&acr=1&ie=utf8&query={n}%ED%9A%8C%EB%A1%9C%EB%98%90')
  html = BeautifulSoup(url.text)

  numbers = html.find('div', class_='num_box').find_all('span')
  del numbers[6]

  box = []

  for i in numbers:
    box.append(int(i.text))
  
  print(f'{n}회차', box)

  time.sleep(1)   # 1초의 간격을 두고 실행
```
